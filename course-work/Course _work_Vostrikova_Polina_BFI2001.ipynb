```python
import math

def task9(n, a): 
    result = math.inf
    c = 1
    mx = 10**10
    while c**(n-1) < mx:
        func = 0
        for i in range(n): func += abs(a[i]-c**i)
        c+=1
        result = min(result, func)
    return result

n = int(input("ะะฒะตะดะธัะต ะบะพะปะธัะตััะฒะพ ัะธัะตะป: "))
a = sorted(map(int,input("ะะฒะตะดะธัะต ัะธัะปะฐ: ").split()))
print(task9(n, a))
```
