```python
import copy
import time
import random

#ввод значений
n = input('Введите кол-во элементов вектора n: ')
if n == '':
    n = 15
min_lim = input('Введите минмальное значения для генерируемого числа min_lim: ')
if min_lim == '':
    min_lim = -100
max_limit = input('Введите максимальное значения для генерируемого числа max_limit: ')
if max_limit == '':
    max_limit = 100

#формирование вектора
array  = [0]*int(n)
for i in range(int(n)):
    array[i] = random.randint(int(min_lim), int(max_limit))

#вывод        
print('Итоговый вектор: ', array) 
```

```python
# Бинарный поиск
# Сравнивает средний элемент отсортированного массива с элементом, который он ищет, чтобы решить, где продолжить поиск
def binary_search(array, element):
    array.sort()
    print ("Отсортированный вектор: ", array)
    print()
    mid = 0
    start = 0
    end = len(array)-1

    while (start <= end):
        mid = (start + end) // 2
        if element == array[mid]: return mid
        elif element < array[mid]: end = mid - 1
        else: start = mid + 1
    return -1

start_time = time.perf_counter()    
    
binary_array = copy.copy(array)
print ("Изначальный вектор: ", array)
element = int(input('Пожалуйста, введите элемент, который необходимо найти: '))
print()
rez = binary_search(binary_array, element)
print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))

if rez != -1:
    print("Номер искомого элемента: ", rez+1)
    question = input("Хотите удалить его? (Да/Нет): ")
    print()
    if (question == "Да") | (question == "да"):
        binary_array.pop(rez)
        print("Итоговый вектор: ", binary_array)
    else: print("Итоговый вектор: ", binary_array)
        
else: 
    question = input("Данного элемента нет в векторе. Хотите добави его? (Да/Нет): ") 
    print()
    if (question == "Да") | (question == "да"):
        index = int(input('Пожалуйста, введите номер места для вставки элемента: '))
        binary_array.insert(index-1, element)
        print("Итоговый вектор: ", binary_array)
    else: print("Итоговый вектор: ", binary_array)
```

```python
#Бинарное дерево
# Создаем узел
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        
def insert(node, key):
    if node is None:
        return Node(key)

    if key < node.key:
        node.left = insert(node.left, key)
    else:
        node.right = insert(node.right, key)
    return node

# Отсортированный обход
def inorder(root):
    if root is not None:
        inorder(root.left)
        # Обходим корень
        print(str(root.key) + " ->", end=' ')
        inorder(root.right)
        
        
def findval(root, element):
    if element < root.key:
        if root.left is None:
            return "Элемент не найден."
        return findval(root.left, element)
    elif element > root.key:
        if root.right is None:
            return "Элемент не найден."
        return findval(root.right, element)
    else:
        return "Элемент найден"


# Поиск inorder-преемника
def minValueNode(node):
    current = node
    # Найдем крайний левый лист — он и будет inorder-преемником
    while(current.left is not None):
        current = current.left
    return current

# Удаление узла
def deleteNode(root, key):
    # Возвращаем, если дерево пустое
    if root is None:
        return root

    # Посик нужного узла
    if key < root.key:
        root.left = deleteNode(root.left, key)
    elif(key > root.key):
        root.right = deleteNode(root.right, key)
    else:
        # Если один дочерний узел или вообще их нет        
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp

        # Если у узла два дочерних узла, помещаем центрированного преемника на место узла, который нужно удалить
        temp = minValueNode(root.right)
        root.key = temp.key
        # Удаляение
        root.right = deleteNode(root.right, temp.key)

    return root



start_time = time.perf_counter()    

binary_t = copy.copy(array)
root = None
for i in range (len(binary_t)):
    root = insert(root, binary_t[i])

print("Отсортированный обход: ", end=' ')
inorder(root)
print()

element = int(input('Пожалуйста, введите элемент, который необходимо найти: '))
q = findval(root, element)
print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))

if q == "Элемент не найден.":
    question = input("Хотите добави его? (Да/Нет): ")
    if (question == "Да") | (question == "да"):
        root = insert(root, element)
        print("Отсортированный обход: ", end=' ')
        inorder(root)
        print()
    else:
        print("Отсортированный обход: ", end=' ')
        inorder(root)
        print()
else: 
    question = input("Хотите удалить его? (Да/Нет): ")
    print()
    if (question == "Да") | (question == "да"):
        root = deleteNode(root, element)
        print("Отсортированный обход: ", end=' ')
        inorder(root)
    else:
        print("Отсортированный обход: ", end=' ')
        inorder(root)
        print()
```

```python
# Метод Фибоначчи
# Использует числа Фибоначчи для вычисления размера блока или диапазона поиска на каждом шаге
def fibonacci_method(array, element):
    array.sort()
    print ("Отсортированный вектор: ", array)
    print()
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    while (fibM < len(array)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1;
    while (fibM > 1):
        i = min(index + fibM_minus_2, (len(array)-1))
        if (array[i] < element):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
        elif (array[i] > element):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        else :
            return i
    if(fibM_minus_1 and index < (len(array)-1) and array[index+1] == element):
        return index+1;
    return -1

start_time = time.perf_counter() 

fibonacci_array = copy.copy(array)
print ("Изначальный вектор: ", array)
element = int(input('Пожалуйста, введите элемент, который необходимо найти: '))
print()
rez = fibonacci_method(fibonacci_array, element)
print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))

if rez != -1:
    print("Номер искомого элемента: ", rez+1)
    question = input("Хотите удалить его? (Да/Нет): ")
    print()
    if (question == "Да") | (question == "да"):
        fibonacci_array.pop(rez)
        print("Итоговый вектор: ", fibonacci_array)
    else: print("Итоговый вектор: ", fibonacci_array)
        
else: 
    question = input("Данного элемента нет в векторе. Хотите добави его? (Да/Нет): ") 
    print()
    if (question == "Да") | (question == "да"):
        index = int(input('Пожалуйста, введите номер места для вставки элемента: '))
        fibonacci_array.insert(index-1, element)
        print("Итоговый вектор: ", fibonacci_array)
    else: print("Итоговый вектор: ", fibonacci_array)
```

```python
# Интерполяционный
# Вычисляет вероятную позицию искомого элемента по формуле
def interpolation_search(array, element):
    array.sort()
    print ("Отсортированный вектор: ", array)
    print()
    start = 0
    end = (len(array) - 1)
    while start <= end and element >= array[start] and element <= array[end]:
        index = start + int(((float(end - start) / ( array[end] - array[start])) * ( element - array[start])))
        if array[index] == element: return index
        if array[index] < element: start = index + 1;
        else: end = index - 1;
    return -1
    
start_time = time.perf_counter()    
    
interpolation_array = copy.copy(array)
print ("Изначальный вектор: ", array)
element = int(input('Пожалуйста, введите элемент, который необходимо найти: '))
print()
rez = interpolation_search(interpolation_array, element)
print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))

if rez != -1:
    print("Номер искомого элемента: ", rez+1)
    question = input("Хотите удалить его? (Да/Нет): ")
    print()
    if (question == "Да") | (question == "да"):
        interpolation_array.pop(rez)
        print("Итоговый вектор: ", interpolation_array)
    else: print("Итоговый вектор: ", interpolation_array)
        
else: 
    question = input("Данного элемента нет в векторе. Хотите добави его? (Да/Нет): ") 
    print()
    if (question == "Да") | (question == "да"):
        index = int(input('Пожалуйста, введите номер места для вставки элемента: '))
        interpolation_array.insert(index-1, element)
        print("Итоговый вектор: ", interpolation_array)
    else: print("Итоговый вектор: ", interpolation_array)
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
board = [[0 for i in range(8)] for j in range(8)] # сама доска

# устанавливает ферзя в клетку [i, j]
def set_queen(i, j):
    for x in range(8):
        board[x][j] += 1
        board[i][x] += 1
        if 0 <= i + j - x < 8:
            board[i + j - x][x] += 1
        if 0 <= i - j + x < 8:
            board[i - j + x][x] += 1
    board[i][j] = -1

#Удаление ферзя
def remove_queen(i, j):
    for x in range(8):
        board[x][j] -= 1
        board[i][x] -= 1
        if 0 <= i + j - x < 8:
            board[i + j - x][x] -= 1
        if 0 <= i - j + x < 8:
            board[i - j + x][x] -= 1
    board[i][j] = 0
    
#Вывод решения
def print_position():
    abc = 'abcdefgh'
    ans = []
    for i in range(8):
        for j in range(8):
            if board[i][j] == -1:
                ans.append(abc[j] + str(i + 1))
    print(', '.join(ans))
    
# решение для i-ой строки
def solve(i): 
    for j in range(8):
        if board[i][j] == 0:
            set_queen(i, j)
            if i == 7:
                print_position()
            else:
                solve(i+ 1) 
            remove_queen(i, j)
                
solve(0)
```
